<h2> Lets see if we can prepredict how many special requests a customer might make based off their booking info</h2>

```python
interested = ["customer_type","total_of_special_requests", "days_in_waiting_list", \
              "booking_changes","reserved_room_type","adults", "children", "babies",\
              "stays_in_weekend_nights", "stays_in_week_nights", "is_repeated_guest", \
              "previous_cancellations", "booking_changes", "lead_time", "country", "distribution_channel"]

clean_df = df[interested]
```

```python
clean_df.sample(5)
```

<p> Some words above such as customer type and reserved room type - need to convert that to numerical representation for it to work withon our ml models</p>

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
clean_df = clean_df.copy()
clean_df.loc[:,"customer_type"] = le.fit_transform(clean_df["customer_type"])
clean_df.loc[:,"distribution_channel"] = le.fit_transform(clean_df["distribution_channel"])
clean_df.loc[:,"country"] = le.fit_transform(clean_df["country"])
clean_df.loc[:,"reserved_room_type"] = le.fit_transform(clean_df["reserved_room_type"])
```

```python
clean_df.head()
```

```python
clean_df.head()
```

```python
y = clean_df["total_of_special_requests"]
X = clean_df.drop("total_of_special_requests", 1)
X = X.fillna(0)
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt

scaler = StandardScaler()
scaled_features = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3) 
clusters = kmeans.fit_predict(scaled_features)
clustered_X = X.copy()
clustered_X['cluster'] = clusters

pca = PCA(n_components=2) # change to 2d
reduced_features = pca.fit_transform(scaled_features)

plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=clusters, cmap='viridis')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('Customer Clusters')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
scaled_features.shape
```

```python
clean_df.columns
```

```python
scaled_mini = scaled_features[:2000,:]
mini_features = pca.fit_transform(scaled_mini)
```

```python
scaled_mini.shape
```

```python
from sklearn.cluster import MeanShift

mean_shift = MeanShift(bandwidth=2)  # bandwidth can be automatically detected using the `estimate_bandwidth` function from sklearn.
clusters_mean_shift = mean_shift.fit_predict(scaled_mini)
```

```python
plt.scatter(mini_features[:, 0], mini_features[:, 1], c=clusters_mean_shift, cmap='viridis')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('Mean Shift Clusters')
plt.colorbar(label='Cluster Label')
plt.show()

```

```python
from sklearn.cluster import AffinityPropagation

affinity_propagation = AffinityPropagation(random_state=42)
clusters_affinity = affinity_propagation.fit_predict(mini_features)
```

```python
plt.scatter(mini_features[:, 0], mini_features[:, 1], c=clusters_affinity, cmap='viridis')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('Affinity Propagation Clusters')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
rd_clf = RandomForestClassifier()
rd_clf.fit(X_train, y_train)

y_pred_rd_clf = rd_clf.predict(X_test)

acc_rd_clf = accuracy_score(y_test, y_pred_rd_clf)
conf = confusion_matrix(y_test, y_pred_rd_clf)
clf_report = classification_report(y_test, y_pred_rd_clf)

print(f"Accuracy Score of Random Forest is : {acc_rd_clf}")
print(f"Confusion Matrix : \n{conf}")
print(f"Classification Report : \n{clf_report}")
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)

acc_knn = accuracy_score(y_test, y_pred_knn)
conf = confusion_matrix(y_test, y_pred_knn)
clf_report = classification_report(y_test, y_pred_knn)

print(f"Accuracy Score of KNN is : {acc_knn}")
print(f"Confusion Matrix : \n{conf}")
print(f"Classification Report : \n{clf_report}")
```
